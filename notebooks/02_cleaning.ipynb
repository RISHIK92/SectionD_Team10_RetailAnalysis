## **Datatype Fixing**

```python
# ID Columns
df['Transaction_ID'] = df['Transaction_ID'].astype('Int64')
df['Customer_ID'] = df['Customer_ID'].astype('Int64')

print(df[['Transaction_ID', 'Customer_ID']].dtypes)
print(df[['Transaction_ID', 'Customer_ID']].isnull().sum())
```

```python
# Date Column
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()

print("Date dtype :", df['Date'].dtype)
print("Year dtype :", df['Year'].dtype)
print("\nAll months present:")
print(df['Month'].value_counts())
```

```python
# Time Column
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce').dt.time

print(df['Time'].dtype)
print(df['Time'].isnull().sum())
print(df['Time'].head(5))
```

```python
# Fix numerical Column
df['Age'] = df['Age'].astype('Int64')
df['Year'] = df['Year'].astype('Int64')
df['Total_Purchases'] = df['Total_Purchases'].astype('Int64')
df['Ratings'] = df['Ratings'].astype('Int64')
```

```python
# Roundoff float Column
df['Amount'] = df['Amount'].round(2)
df['Total_Amount'] = df['Total_Amount'].round(2)
```

```python
# Categorical Column
cat_cols = [
    'City','State','Country','Gender','Income','Customer_Segment',
    'Product_Category','Product_Brand','Product_Type',
    'Feedback','Shipping_Method','Payment_Method','Order_Status','products'
]

for col in cat_cols:
    df[col] = df[col].astype('category')
```

```python
# Month Column
df['Month'] = pd.Categorical(
    df['Month'],
    categories=[
        'January','February','March','April','May','June',
        'July','August','September','October','November','December'
    ],
    ordered=True
)
```

### **Handle missing values**

```python
# Check null values
df.isnull().sum()
```

```python
# Drop rows
df = df.dropna(subset=['Transaction_ID','Customer_ID','Amount'])
```

```python
# Numerical Imputation
num_cols = ['Age','Ratings']

for col in num_cols:
   df[col].fillna(df[col].median(), inplace=True)
```

```python
# Categorical Imputation
cat_cols = ['Gender','Customer_Segment','Product_Category',
            'Product_Brand','Feedback','Shipping_Method',
            'Payment_Method','Order_Status','City','State','Country','Income']

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# Handle missing values Total_Purchases, Amount, Total_Amount
df[df[['Total_Purchases', 'Amount', 'Total_Amount']].isnull().any(axis=1)]
```

```python
# Ensure numeric
cols = ['Total_Purchases', 'Amount', 'Total_Amount']
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')

# Convert purchases to float for calculation
df['Total_Purchases'] = df['Total_Purchases'].astype('float64')

# Fill missing values

# Case 1: Fill Total_Purchases
mask = df['Total_Purchases'].isna() & df['Total_Amount'].notna() & df['Amount'].notna() & (df['Amount'] != 0)
df.loc[mask, 'Total_Purchases'] = df.loc[mask, 'Total_Amount'] / df.loc[mask, 'Amount']

# Case 2: Fill Amount
mask = df['Amount'].isna() & df['Total_Amount'].notna() & df['Total_Purchases'].notna() & (df['Total_Purchases'] != 0)
df.loc[mask, 'Amount'] = df.loc[mask, 'Total_Amount'] / df.loc[mask, 'Total_Purchases']

# Case 3: If both missing → use median
df['Amount'].fillna(df['Amount'].median(), inplace=True)
df['Total_Purchases'].fillna(df['Total_Purchases'].median(), inplace=True)

# FINAL CONSISTENCY FIX

# Round purchases to integer
df['Total_Purchases'] = df['Total_Purchases'].round().astype('Int64')

# Recalculate Total_Amount
df['Amount'] = df['Amount'].round(2)
df['Total_Amount'] = (df['Amount'] * df['Total_Purchases']).round(2)
```

```python
# Fix Datetime columns
df = df.dropna(subset=['Date'])

df['Year'] = df['Date'].dt.year.astype('Int64')
df['Month'] = df['Date'].dt.month_name()

df['Time'] = df['Time'].fillna('00:00:00')
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce').dt.time
```
